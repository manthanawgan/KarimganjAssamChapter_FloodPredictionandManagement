```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/merged_data.csv')
df.head()
```

```python
#Define new column names
new_column_names = {
    "Date": "date",
    "temperature_2m_max (°C)": "temp_max",
    "temperature_2m_min (°C)": "temp_min",
    "temperature_2m_mean (°C)": "temp_mean",
    "rain_sum (mm)": "rain_mm",
    "precipitation_hours (h)": "precip_hours",
    "wind_speed_10m_max (m/s)": "wind_max",
    "wind_gusts_10m_max (m/s)": "wind_gusts",
    "wind_direction_10m_dominant (°)": "wind_dir",
    "et0_fao_evapotranspiration (mm)": "evapotranspiration",
    "unknown_discharge (m³/s)": "discharge_unknown",
    "Kushi_discharge (m³/s)": "discharge_kushi",
    "Longai_discharge (m³/s)": "discharge_longai",
    "Singla_discharge (m³/s)": "discharge_singla",
    "pressure_msl (hPa)": "pressure",
    "soil_moisture_0_to_7cm (m³/m³)": "soil_moisture_0_7",
    "soil_moisture_7_to_28cm (m³/m³)": "soil_moisture_7_28",
    "soil_moisture_28_to_100cm (m³/m³)": "soil_moisture_28_100",
    "soil_moisture_100_to_255cm (m³/m³)": "soil_moisture_100_255",
    "flooded": "flooded"
}


df.rename(columns=new_column_names, inplace=True)
```

```python
#Save with new column names
df.to_csv("cleaned_data.csv", index=False)
print("Column names updated successfully!")
```
